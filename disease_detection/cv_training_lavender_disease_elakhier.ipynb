```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
%matplotlib inline


# Load a model

model = YOLO("yolo11x.pt")  # load a pretrained model 

# Train the model
results = model.train(data="dataset/data.yaml", epochs=100,batch = 1, imgsz=640,device='cuda:0') #batch2
# Extract training history
train_loss = results.metrics['train/box_loss']
val_loss = results.metrics['val/box_loss']
accuracy = results.metrics['metrics/precision']

# Plot Loss Curves
plt.figure(figsize=(12, 6))
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title("Training and Validation Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()

# Plot Accuracy Curves
plt.figure(figsize=(12, 6))
plt.plot(accuracy, label='Validation Precision')
plt.title("Model Precision Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Precision")
plt.legend()
plt.grid()
plt.show()
```

```python
import torch  # PyTorch for model loading
from ultralytics import YOLO
import cv2  # OpenCV for image processing
import matplotlib.pyplot as plt  # For displaying the image

# Ensure matplotlib plots are shown inline
%matplotlib inline

# Load the YOLO model
model = YOLO('runs/detect/train20/weights/best.pt')  # Path to your trained model

# Load an image
image_path = 'new_images/Brown-leaves-of-lavender-.webp'  # Replace with your image file path
img = cv2.imread(image_path)

# Convert BGR image to RGB (YOLO expects RGB format)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perform inference
results = model(img_rgb)

# Count the number of bounding boxes and calculate accuracy
num_bounding_boxes = len(results[0].boxes)  # Access the first result's bounding boxes
if num_bounding_boxes > 0:
    confidences = [box.conf.item() for box in results[0].boxes]  # Extract confidence scores
    average_confidence = sum(confidences) / len(confidences)  # Calculate average confidence
else:
    average_confidence = 0  # If no detections, accuracy is 0

print(f"Number of bounding boxes: {num_bounding_boxes}")
print(f"Average confidence (accuracy): {average_confidence:.2f}")

# Draw bounding boxes on the image
for box in results[0].boxes:
    # Extract box coordinates and convert to integer
    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
    # Draw the box on the original image
    cv2.rectangle(img_rgb, (x1, y1), (x2, y2), color=(255, 0, 0), thickness=2)

# Display the image with bounding boxes and accuracy
plt.figure(figsize=(10, 10))
plt.imshow(img_rgb)
plt.axis('off')  # Turn off the axis
plt.title(f"Detected {num_bounding_boxes} Bounding Boxes | Accuracy: {average_confidence:.2%}")
plt.show()
```
